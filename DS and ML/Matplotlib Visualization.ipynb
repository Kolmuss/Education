```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data
data = np.linspace(0, 9, 100)

# Plot the data
plt.plot(data, data, label='linear')
# Add label text
plt.ylabel('some numbers')
plt.xlabel('some numbers')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

<h3>TASK 2: Creating graphs </h3>

<b>Task 2.1</b> Plotting bar charts
<br/>
On the example below, the bar chart for Seasons was plotted the documentation [here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)

```python
plt.bar(data.Season.unique(), data.groupby('Season')['Sales'].sum(), width = 0.9, 
        color='#64fd00', ec='#414142')
plt.xlabel('Seasons')
plt.ylabel('Sales amount')
plt.title('Sales over seasons')
plt.show()
```

<b>Now, plot a bar char for Days and Revenue means </b>

```python
plt.bar(data['Day'].unique(), data.groupby('Day')['Revenue'].mean(), width = 0.4, 
        color='blue', ec='black') 
plt.xlabel('Days of week')
plt.xticks(rotation=90)
plt.ylabel('Agerage revenue')
plt.title('Days and Revenue')
plt.show()
```

<b>Task 2.2</b> Plotting bar charts
<br/> Plot a bar chart visualizing <b>Locations and Sales </b>

```python
# Your code goes here
plt.bar(data['Location'].unique(), data.groupby('Location')['Sales'].sum(),
       color='red', ec='black')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('Sales')
```

<b>Task 2.3</b> Plotting bar charts
<br/> Plot two bar charts visualizing <b>Any month days + Sales </b> and <b> June days + Sales</b>

```python
# Your code goes here
space, (sub1, sub2) = plt.subplots(1, 2, sharey=True, figsize=(50,20))
sub1.bar(list(np.arange(1,32)), data[data['Date'].dt.month == 7]['Sales'])
sub2.bar(list(np.arange(1,31)), data[data['Date'].dt.month == 6]['Sales'])
space.show()
```

<b>TASK 3</b> Scatterplots

Take a look at well known <u>Temperature - Sales relationship</u>, visualized by scatterplot

```python
plt.scatter(data.Temperature, data.Sales, 50, marker = '*', color = '#2Ffd50')
plt.yticks(np.arange(0, 60, 5))
plt.xticks(np.arange(10, 130, 10))
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.axis('equal')
plt.title('Sales and Temperature relationship')
plt.show()
```

<b> Task 3.1 </b> Create a scatterplot for <u>Revenue - Flyers </u>, use legend method of scatterplot

```python
# Your code goes here
plt.scatter(data.Revenue, data.Flyers, marker = '+', color = 'red', ec='black')

plt.xlabel('Revenue')
plt.ylabel('Flyers')

plt.title('Revenue and Flyers relationship')
plt.show()
```

<b>Task 3.2 </b> Create a scatterplot for <u> Any month's days - Sales </u>

```python
plt.scatter(list(np.arange(1,32)), data[data['Date'].dt.month == 7]['Sales'], 50, marker = '+', color = 'red', ec='black')

plt.xlabel('June')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.title('Sales for June')
plt.show()
```
