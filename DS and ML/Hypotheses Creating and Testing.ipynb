<span style="color:#011f4b;"><b>A <code>Confidence Interval</code></b> is a range of values we are fairly sure our true value lies in.The interval has an associated <code>confidence level</code> that the true parameter is in the proposed range. In general terms, a confidence interval for an unknown parameter is based on sampling the distribution of a corresponding estimator. For example, if the <b>confidence level (CL) is 90%</b> then in hypothetical indefinite data collection, <b>in 90% of the samples the interval estimate will contain the true population parameter</b>.
    <br/><br/>
    Most commonly, <code>the 95%</code> confidence level is used. However, confidence levels of 90% and 99% are also often used in analys. 
    <br/><br/>
    Factors affecting the width of the confidence interval include the size of the sample, the confidence level, and the variability in the sample. 
    <br/>A larger sample will tend to produce a better estimate of the population parameter, when all other factors are equal. A higher confidence level will tend to produce a broader confidence interval.  
</span>


<span style="color:#011f4b;"><code>Hypothesis testing</code> in statistics is a way for you to test the results of a survey or experiment to see if you have meaningful results. Youre basically testing whether your results are valid by figuring out the odds that your results have happened by chance. If your results may have happened by chance, the experiment wont be repeatable and so has little use.
    <br/><br/>
    In Hypothesis Testing you operate with terms like <code>null hypothesis</code> and <code>alternative hyphotesis</code>, <code>level of significance</code>, <code> one-side and two-side test</code>.
    <br/><br/>
    <b>Null hypothesis:</b> Null hypothesis is a statistical hypothesis that assumes that the observation is due to a chance factor.  Null hypothesis is denoted by; H0: 1 = 2, which shows that there is no difference between the two population means.
    <br/><br/>
    <b>Alternative hypothesis: </b>Contrary to the null hypothesis, the alternative hypothesis shows that observations are the result of a real effect.
    <br/> <br/>
    <b>Level of significance:</b> Refers to the degree of significance in which we accept or reject the null-hypothesis.  100% accuracy is not possible for accepting or rejecting a hypothesis, so we therefore select a level of significance that is usually 5%.
 <br/><br/>
<b>One-side test:</b> When the given statistical hypothesis is one value like H0: 1 = 2, it is called the one-tailed test.
<br/>
<b>Two-side test:</b> When the given statistics hypothesis assumes a less than or greater than value, it is called the two-tailed test.
</span>

<b>Task 2.1</b> Calculate CI for specified values (provided by your instructor) with with given CL (provided by your instructor). You are prohibited to use ready-to-use packages

```python
# Your code goes here
df['G1']
```

```python
def Z_dist(sample):
    std = sample.std()
    mean = sample.mean()
    sample = (sample - mean)/std
    return sample
```

```python
def CI(sample, CL):
    if (CL == .75):
        Z = 1.15
    elif(CL == .85):
        Z = 1.44
    elif(CL == .95):
        Z = 1.96
    dist = sample
    st = dist.mean()-dist.std()*Z
    end = dist.mean()+dist.std()*Z
    return (st,end)
```

```python
p = ['G1', 'G2', 'G3']
CLs = [.75,.85,.95]
for c in p:
    start = []
    end = []
    for i in CLs:
        start.append(CI(df[c],i)[0])
        end.append(CI(df[c],i)[1]-CI(df[c],i)[0])
    N = 3
    ind = np.arange(N)  
    width = 0.05
    plt.figure(figsize=(5, 5))
    p1 = plt.bar(ind, start, width, color='white')
    p2 = plt.bar(ind, end, width,
                 bottom=start)
    plt.xticks(ind, (.75, .85, .95))
    plt.title(c)
    plt.show()
```

<b>Task 2.1 </b> Calculate CI for proportions (specified by instructor) with given CLs. You are prohibited to use ready-to-use packages. 

```python
def CI_prop(sample, CL):
    if (CL == .8):
        Z = 1.282
    elif(CL == .9):
        Z = 1.645
    elif(CL == .99):
        Z = 2.576
    p = sample.value_counts(normalize=True)[0]
    return (p-Z*np.sqrt(p*(1-p)/sample.shape[0]), p+Z*np.sqrt(p*(1-p)/sample.shape[0]))
```

```python
p = ['Pstatus', 'famsize', 'address']
CLs = [.80,.90,.99]
for c in p:
    start = []
    end = []
    for i in CLs:
        start.append(CI_prop(df[c],i)[0])
        end.append(CI_prop(df[c],i)[1]-CI_prop(df[c],i)[0])
    N = 3
    ind = np.arange(N)  
    width = 0.05
    plt.figure(figsize=(5, 5))
    p1 = plt.bar(ind, start, width, color='white')
    p2 = plt.bar(ind, end, width,
                 bottom=start)
    plt.xticks(ind, (.75, .85, .95))
    plt.title(c)
    plt.show()
```
