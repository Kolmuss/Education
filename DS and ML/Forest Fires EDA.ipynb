<span style="color:#011f4b;"><b>A <code>correlation</code></b> is a statistic intended to quantify the strength of the relationship between two variables. <code>Pearsons correlation</code> is always between -1 and +1 (including both). If <code>r</code> is positive, we say that the correlation is positive, which means that when one
variable is high, the other tends to be high. If <code>r</code> is negative, the correlation is negative, so when one variable is high, the other is low.
    <br/><br/>
    The magnitude of <code>r</code> indicates the strength of the correlation. If <code>r</code> is 1 or -1,
the variables are perfectly correlated, which means that if you know one, you
can make a perfect prediction about the other.
<br/><br/> Most correlation in the real world is not perfect, but it is still useful.
    <br/><br/>If <u>Pearsons correlation is near 0</u>, it is tempting to conclude that there is no
relationship between the variables, but that conclusion <b>is not valid.</b> Pearsons correlation only measures <code>linear relationships</code>. If theres a nonlinear relationship, r understates its strength.
<br/> <br/>
    <code>Pearsons correlation</code> works well if the relationship between variables <u>is linear</u>
    and if the variables are roughly normal. But it is not robust in the presence of <u>outliers</u>. <b><code>Spearmans rank</code></b> correlation is an alternative that mitigates the effect of outliers and skewed distributions.
<br/>
 If the relationship is nonlinear, Pearsons correlation tends to underestimate the strength of the relationship, and
<br/> Pearsons correlation can be affected (in either direction) if one of the distributions is skewed or contains outliers. Spearmans rank correlation is more robust.
    <br/><br/>
<b><span style="font-size:14px; color:#420420">The moral of this story is that you should always look at a scatter plot of
your data before blindly computing a correlation coefficient. </span></b>
    <br/> <br/> <br/>
   <b> <code>Covariance</code></b> is a measure of the tendency of two variables to vary together. Covariance is useful in some computations, but it is seldom reported as a summary statistic because it is hard to interpret. We will not stop on Covariance usage.</span>
